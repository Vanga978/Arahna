```python
# 7.3 все методы объединения join
import pandas as pd
data_1 = pd.DataFrame({'Value': [100, 45, 80],'Group': [1, 4, 5]},index = ['I0', 'I1', 'I2'])
data_2 = pd.DataFrame({'Company': ['Google', 'Amazon', 'Facebook'],'Add': ['S0', 'S1', 'S7']},index = ['I0', 'I1', 'I3'])
data_12I=data_1.join(data_2, how='inner')
data_12O=data_1.join(data_2, how='outer')
data_12L=data_1.join(data_2, how='left')
data_12R=data_1.join(data_2, how='right')
display(data_1)
display(data_2)
display(data_12I)
display(data_12O)
display(data_12L)
display(data_12R)

```
