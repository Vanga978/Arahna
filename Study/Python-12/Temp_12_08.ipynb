```python
# 8.1
import pandas as pd
import re
ratings_movies=pd.read_csv('data/ratings_movies.csv')
display(ratings_movies.head(3))

def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None НО ДЛЯ ПОДСЧЕТА 4444
        return 4444

#c = re.findall(r'\(\d{4}\)',ratings_movies.loc[0]['title']) 

ratings_movies['year_release']=ratings_movies['title'].apply(get_year_release)

display(ratings_movies.head(3))
ratings_movies['year_release'].value_counts()
ratings_movies['year_release'].value_counts().max()
display('----------')
ratings_movies[ratings_movies['year_release'] == 4444].value_counts().shape[0]

```
